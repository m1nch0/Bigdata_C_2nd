````python
cols_numerical = []
for col in df2:
    if df2[col].dtype == 'int64':
        cols_numerical.append(col)
cols_numerical = cols_numerical[1:]
cols_ordinal = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual']
cols_categorical = ['Foundation','Heating', 'CentralAir', 'Electrical']
````

## 이상치 처리
모델링 후 진행

## 결측치 처리


````python
df2[cols_numerical].info()
df2[cols_categorical].info()
df2[cols_ordinal].info()
````

````python
# numerical은 결측치 없음
# categorical은 결측치를 하나의 unique갑으로 보고 0에 대응시킴
df2[cols_categorical] = df2[cols_categorical].fillna(0)
# ordinal은 data description에 따라 NA = 0인 경우 0 처리
df2[cols_ordinal[:4]] = df2[cols_ordinal[:4]].fillna(0)
# 설명이 없을 경우 0.5로 처리
df2[cols_ordinal[4:]] = df2[cols_ordinal[4:]].fillna(0.5)
````

## 수치화

````python
# 카테고리형
for col in df2[cols_categorical]:
    val = 1
    temp_dict = {}
    for unique in df2[col].unique():
        if unique != 0:
            temp_dict[unique] = val
            df2[col][df2[col] == unique] = val
            val += 1
    print(temp_dict)
# df2.loc[:,['Foundation','Heating', 'CentralAir', 'Electrical']]
````

````python
# 서수형
df_temp = df2[['BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual']]
for col in df_temp:    
    df2[col][df2[col] == 'Ex'] = (5 - 1/2) / 5
    df2[col][df2[col] == 'Gd'] = (4 - 1/2) / 5
    df2[col][df2[col] == 'TA'] = (3 - 1/2) / 5
    df2[col][df2[col] == 'Fa'] = (2 - 1/2) / 5
    df2[col][df2[col] == 'Po'] = (1 - 1/2) / 5
    print(df2[col].unique())
````

````python
df2['BsmtExposure'].unique()

df2['BsmtExposure'][df2['BsmtExposure']=='Gd'] = (4 - 1/2) / 4
df2['BsmtExposure'][df2['BsmtExposure']=='Av'] = (3 - 1/2) / 4
df2['BsmtExposure'][df2['BsmtExposure']=='Mn'] = (2 - 1/2) / 4
df2['BsmtExposure'][df2['BsmtExposure']=='No'] = (1 - 1/2) / 4

df2['BsmtExposure'].unique()
````

````python
df2['BsmtFinType1'].unique()

df2['BsmtFinType1'][df2['BsmtFinType1']=='GLQ'] = (6 - 1/2) / 6
df2['BsmtFinType1'][df2['BsmtFinType1']=='ALQ'] = (5 - 1/2) / 6
df2['BsmtFinType1'][df2['BsmtFinType1']=='BLQ'] = (4 - 1/2) / 6
df2['BsmtFinType1'][df2['BsmtFinType1']=='Rec'] = (3 - 1/2) / 4
df2['BsmtFinType1'][df2['BsmtFinType1']=='LwQ'] = (2 - 1/2) / 4
df2['BsmtFinType1'][df2['BsmtFinType1']=='Unf'] = (1 - 1/2) / 4

df2['BsmtFinType1'].unique()
````

````python
df2['BsmtFinType2'].unique()

df2['BsmtFinType2'][df2['BsmtFinType2']=='GLQ'] = (6 - 1/2) / 6
df2['BsmtFinType2'][df2['BsmtFinType2']=='ALQ'] = (5 - 1/2) / 6
df2['BsmtFinType2'][df2['BsmtFinType2']=='BLQ'] = (4 - 1/2) / 6
df2['BsmtFinType2'][df2['BsmtFinType2']=='Rec'] = (3 - 1/2) / 4
df2['BsmtFinType2'][df2['BsmtFinType2']=='LwQ'] = (2 - 1/2) / 4
df2['BsmtFinType2'][df2['BsmtFinType2']=='Unf'] = (1 - 1/2) / 4

df2['BsmtFinType2'].unique()
````

## 표준화
정확히는 Standardization: (요소값 - 평균)/표준편차 <p>
Normalization: (요소값 - 최소값)/(최대값 - 최소값)


````python
# 수치형 데이터
for col in cols_numerical:
    df2[col] = (df2[col] - df2[col].mean()) / df2[col].std()
df2[cols_numerical].head(10)
````
