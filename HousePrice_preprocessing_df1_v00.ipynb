```python
cols_num = ['LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea']
cols_ord = ['Street', 'Alley', 'LotShape', 'Utilities', 'LandSlope', 'OverallQual', 'OverallCond', 'ExterQual']
cols_cat = ['MSSubClass', 'MSZoning', 'LandContour', 'LotConfig', 'Neighborhood', 'Condition1', 'Condition2',
            'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterCond']
```

### 결측값 (nan) 처리
    - numerical (수치형) : LotFrontage(거리에서 집까지 발걸음), MasVnrArea(벽돌장식 면적)  
    - ordinal (서수형) : Alley(골목길 접근성)  
    - categorical (카테고리형) : MasVnrType(벽돌장식 타입)

```python
df1[cols_num].head()
```

```python
# 수치형 결측값은 해당 컬럼의 평균값 입력
df1[cols_num[0]] = df1[cols_num[0]].fillna(df1['LotFrontage'].mean())
df1[cols_num[4]] = df1[cols_num[4]].fillna(df1['MasVnrArea'].mean())
```

```python
df1[cols_ord].head()
```

```python
# 서수형 결측값은 data description에 따라 NA = 0인 경우 0으로 입력
df1[cols_ord[1]] = df1[cols_ord[1]].fillna(0)
```

```python
df1[cols_cat].head()
```

```python
# 카테고리형 결측값은 하나의 unique 값으로 판단하여 0으로 입력
df1[cols_cat[-2]] = df1[cols_cat[-2]].fillna(0)
```

```python
df1.isnull().sum()
```

### 수치화

```python
# 서수형

tmp_ord = df1[['Street', 'Alley']]
for col in tmp_ord:
    df1[col][df1[col] == 'Grvl'] = (1 - 1/2) / 2
    df1[col][df1[col] == 'Pave'] = (2 - 1/2) / 2
    
    print(df1[col].unique())
```

```python
df1['LotShape'].unique()

df1['LotShape'][df1['LotShape'] == 'Reg'] = (4 - 1/2) / 4
df1['LotShape'][df1['LotShape'] == 'IR1'] = (3 - 1/2) / 4
df1['LotShape'][df1['LotShape'] == 'IR2'] = (2 - 1/2) / 4
df1['LotShape'][df1['LotShape'] == 'IR3'] = (1 - 1/2) / 4

df1['LotShape'].unique()
```

```python
df1['Utilities'].unique()

df1['Utilities'][df1['Utilities'] == 'AllPub'] = (4 - 1/2) / 4
df1['Utilities'][df1['Utilities'] == 'NoSewr'] = (3 - 1/2) / 4
df1['Utilities'][df1['Utilities'] == 'NoSeWa'] = (2 - 1/2) / 4
df1['Utilities'][df1['Utilities'] == 'ELO'] = (1 - 1/2) / 4

df1['Utilities'].unique()
```

```python
df1['LandSlope'].unique()

df1['LandSlope'][df1['LandSlope'] == 'Gtl'] = (3 - 1/2) / 3
df1['LandSlope'][df1['LandSlope'] == 'Mod'] = (2 - 1/2) / 3
df1['LandSlope'][df1['LandSlope'] == 'Sev'] = (1 - 1/2) / 3

df1['LandSlope'].unique()
```

```python
df1['ExterQual'].unique()

df1['ExterQual'][df1['ExterQual'] == 'Ex'] = (5 - 1/2) / 5
df1['ExterQual'][df1['ExterQual'] == 'Gd'] = (4 - 1/2) / 5
df1['ExterQual'][df1['ExterQual'] == 'TA'] = (3 - 1/2) / 5
df1['ExterQual'][df1['ExterQual'] == 'Fa'] = (2 - 1/2) / 5
df1['ExterQual'][df1['ExterQual'] == 'Po'] = (1 - 1/2) / 5

df1['ExterQual'].unique()
```

```python
# 카테고리형

for col in df1[cols_cat]:
    val = 1
    tmp_dict = {}
    
    for unique in df1[col].unique():
        
        if unique != 0:
            tmp_dict[unique] = val
            df1[col][df1[col] == unique] = val
            val += 1
            
    print(tmp_dict)
```

```python
df1[cols_cat].head()
```

### 표준화
    Standardization : (요소값 - 평균) / 표준편차
    Normalization : (요소값 - 최소값) / (최대값 - 최소값)

```python
# 수치형

for col in cols_num:
    df1[col] = (df1[col] - df1[col].mean()) / df1[col].std()

df1[cols_num].head()
```
